```python
#Q4)The Phone_Number column has inconsistent formats. What steps can be taken to standardize it?
import pandas as pd
import re

def standardize_phone_number(phone_number):
  """Standardizes phone numbers to a consistent format.

  Args:
    phone_number: The phone number string to standardize.

  Returns:
    The standardized phone number string, or None if the input is invalid.
  """
  if pd.isna(phone_number):
    return None

  phone_number = str(phone_number)


  phone_number = re.sub(r'\D', '', phone_number)

  if len(phone_number) == 10:
    return f"+1 ({phone_number[:3]}) {phone_number[3:6]}-{phone_number[6:]}"
  elif len(phone_number) == 11 and phone_number[0] == '1':
    return f"+1 ({phone_number[1:4]}) {phone_number[4:7]}-{phone_number[7:]}"
  elif 10 < len(phone_number) <= 15:
      return "+" + phone_number
  else:
    return None


df['Phone_Number'] = df['Phone_Number'].apply(standardize_phone_number)


print(df.head())
```
